```python
import pandas as pd

df = pd.read_csv("Financials.csv")
df.columns = df.columns.str.strip()

def clean_currency(col):
    return col.replace('[\$,]', '', regex=True).astype(float)

def clean_profit(value):
    if isinstance(value, str):
        value = value.strip()
        if value == '$-':
            return 0.0
        if '(' in value and ')' in value:
            value = value.replace('(', '-').replace(')', '')
        value = value.replace('$', '').replace(',', '')
    return float(value)

columns_to_clean = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Sales', 'COGS']
for col in columns_to_clean:
    df[col] = clean_currency(df[col])

df['Profit'] = df['Profit'].apply(clean_profit)
df.head()
```

```python
import pandas as pd

# Charger les donnÃ©es
df = pd.read_csv("Financials.csv")
df.columns = df.columns.str.strip()  # Nettoyer les noms de colonnes

# Fonction pour nettoyer les valeurs monÃ©taires
def clean_currency(value):
    # Remplacer les symboles et virgules, puis convertir en float
    if isinstance(value, str):
        value = value.replace('$', '').replace(',', '').strip()
    try:
        return float(value)
    except:
        return pd.NA  # ou np.nan

# Colonnes Ã  convertir (adaptÃ© selon tes besoins)
colonnes_a_nettoyer = ['Profit', 'Sales', 'Gross Sales', 'COGS']

for col in colonnes_a_nettoyer:
    df[col] = df[col].apply(clean_currency)

# VÃ©rifier la conversion en affichant un aperÃ§u
print(df[['Profit', 'Sales', 'Gross Sales']].head())

# Recalculer les ratios en vÃ©rifiant que les valeurs sont numÃ©riques
df['Marge Nette'] = df['Profit'] / df['Sales']
df['ROA'] = df['Profit'] / df['Gross Sales']
df['Rotation Actifs'] = df['Sales'] / df['Gross Sales']

# Afficher un aperÃ§u des colonnes calculÃ©es
print(df[['Marge Nette', 'ROA', 'Rotation Actifs']].head())
```

## K-Means
The K-Means algorithm is used here to group similar data points based on their financial attributes. This unsupervised learning method helps identify hidden patterns and segmentations in the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# On suppose que 'df' est dÃ©jÃ  chargÃ© et nettoyÃ© et que les colonnes nÃ©cessaires y figurent.
# VÃ©rifie que les espaces Ã©ventuels ont Ã©tÃ© enlevÃ©s
df.columns = df.columns.str.strip()

# SÃ©lectionner les ratios pour le clustering
features_cluster = ['Marge Nette', 'ROA', 'Rotation Actifs']

# On Ã©limine les lignes comportant des valeurs manquantes sur ces colonnes
df_cluster = df[features_cluster].dropna()

# Standardiser les donnÃ©es
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_cluster)

# Appliquer K-Means (nombre de clusters ici fixÃ© Ã  3, Ã  ajuster selon tes besoins)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Ajouter les labels de clusters dans le DataFrame (attention, on travaille sur df_cluster)
df_cluster['Cluster'] = clusters

# RÃ©duction de dimension avec PCA (2 composantes)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Affichage des clusters via un scatter plot
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', alpha=0.7)
plt.title("Clustering par K-Means visualisÃ© par PCA")
plt.xlabel("Composante Principale 1")
plt.ylabel("Composante Principale 2")
plt.show()
```

## Random Forest
A Random Forest model is applied for classification or regression tasks. This ensemble method leverages multiple decision trees to improve prediction accuracy and highlight the most important financial features.


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# DÃ©finition de la cible et des variables explicatives
features_rf = ['Sales', 'Gross Sales', 'COGS', 'Marge Nette', 'ROA', 'Rotation Actifs']
target = 'Profit'

# VÃ©rifier que ces colonnes existent et ne contiennent pas de valeurs manquantes
df_rf = df.dropna(subset=features_rf + [target])
X_rf = df_rf[features_rf]
y_rf = df_rf[target]

# Division des donnÃ©es : 80% entraÃ®nement, 20% test
X_train, X_test, y_train, y_test = train_test_split(X_rf, y_rf, test_size=0.2, random_state=42)

# EntraÃ®ner le modÃ¨le Random Forest
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# PrÃ©dire et calculer l'erreur moyenne quadratique (MSE)
y_pred = rf.predict(X_test)
print("MSE du Random Forest :", mean_squared_error(y_test, y_pred))

# Afficher l'importance des variables
print("Importance des variables :")
for feature, importance in zip(features_rf, rf.feature_importances_):
    print(f" - {feature}: {importance:.4f}")
```

## Ridge-Lasso
These regularized regression techniques are used to handle multicollinearity and overfitting. Ridge and Lasso help in selecting the most influential predictors and building robust financial models.

```python
from sklearn.linear_model import Ridge, Lasso

# ModÃ¨le Ridge
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
print("MSE du modÃ¨le Ridge :", mse_ridge)
print("Coefficients du modÃ¨le Ridge :")
for feature, coef in zip(features_rf, ridge.coef_):
    print(f" - {feature}: {coef:.4f}")

# ModÃ¨le Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
print("MSE du modÃ¨le Lasso :", mse_lasso)
print("Coefficients du modÃ¨le Lasso :")
for feature, coef in zip(features_rf, lasso.coef_):
    print(f" - {feature}: {coef:.4f}")
```

PCA is employed to reduce the dimensionality of the dataset while preserving as much variance as possible. It simplifies the dataset and enhances visualization and modeling by projecting the data into a smaller set of uncorrelated variables.

**Economic Interpretation:**

Sales are generally the most influential factor in financial performance. They reflect a companyâ€™s fundamental ability to generate revenue. In practice, this underscores the critical importance of market share and sales volume as core drivers of success.

Gross Sales represent total revenue before deductions. This highlights how operational scale directly impacts overall profitability. Companies that reach critical mass often benefit from favorable economies of scale.

Net Margin measures profit efficiency. It reflects cost control and pricing power. A high net margin often signals a sustainable competitive advantage.

COGS (Cost of Goods Sold) indicates direct production costs. Analyzing this metric helps assess production efficiency and supply chain management. Tight control over COGS is essential to protect margins.

ROA measures how effectively a company uses its assets to generate profits. A high ROA indicates good resource allocation and capital efficiency.

Asset Turnover quantifies how much revenue is generated

per dollar of assets. It reflects operational efficiency and how well a company converts investment into sales. High turnover can compensate for lower margins.
